```python
terror.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                       'provstate': 'State', 'city': 'City', 'attacktype1_txt':'AttackType','target1':'Target',
                       'nkill':'Killed', 'nwound':'Wounded','targtype1_txt':'TargetType', 
                       'natlty1_txt': 'TargetNationality', 'ransom': 'Ransom','weaptype1_txt':'Weapon_type',
                       'weapsubtype1_txt': 'WeaponSubtype', 'gname':'Group',
                       'motive':'Motive', 'summary':'Summary', 'addnotes': 'Notes', 'scite1':'Cite1',
                      'scite2':'Cite2', 'scite3':'Cite3'},inplace=True)
```

```python
terror_eda = terror
terror_eda["Frequency"] = terror.groupby(['Year','Country'])['eventid'].transform('count')
```

```python
fig = px.line(terror_eda, x="Year", y="Frequency",line_group="Country", color="Region",
              title='Global Terrorism')
fig.update_xaxes(
    dtick="M1",
    tickformat="Y")
# fig.update_layout(
#     autosize=False,
#     width=1500,
#     height=800,)
fig.show()
```

```python
terror_freq = terror_eda.groupby(['Year', 'Country']).size().reset_index()
terror_freq.rename(columns={0: 'Frequency'},inplace=True)
# terror_freq.sort_values('Frequency', ascending = False, inplace=True)
# terror_freq
```

```python
full_df = pd.merge(terror, humanFreedom, on=['Country', 'Year'] ,how='outer')
full_df['Frequency'].fillna(0, inplace = True)
```

```python
terror_by_country = full_df.groupby(['Country']).size().reset_index()
terror_by_country.rename(columns={0: 'Frequency'},inplace=True)
# terror_by_country.sort_values('Frequency', ascending = False, inplace=True)
# terror_by_country
```

```python
fig = px.choropleth(terror_by_country, locations="Country", locationmode="country names",
                    color="Frequency", title='Global Terrorism',
                    hover_name="Country", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Plasma)

fig.show()
```

```python
unique_years = list(terror_freq['Year'].unique())
recent_years = [year for year in unique_years if year > 2007]
top_10 = dict()
top_10_countries = []
for year in recent_years:
  top_10[year] = terror_freq.loc[terror_freq["Year"] == year].sort_values('Frequency', ascending = False)[:10]
  top_10_countries.extend(top_10[year]["Country"].values)
top_10_countries = list(set(top_10_countries))
```

```python
for _, v in top_10.items():
  print(v)
  print("\n")
```

```python
df = pd.merge(terror_freq, humanFreedom, on=['Country', 'Year'] ,how='right')
df['Frequency'].fillna(value=0, inplace = True)
df.head(5)
```

```python
df_top_10 = df.loc[df["Country"].isin(top_10_countries)]
df_top_10.head(3)
```

```python
def search_top_10(country):
  country = country.upper()
  found = []
  for _, v in top_10.items():
    if not v[v['Country'] == country].empty:
      found.append(v[v['Country'] == country])
  
  for e in found:
    print(e)
    print("\n")
```

```python
search_top_10("libya")
```

```python
fig = px.line(df_top_10, x="Year", y="CivilJustice",line_group="Country", color="Country",
              title='Global Terrorism')
fig.update_xaxes(
    dtick="M1",
    tickformat="Y")
fig.show()
```
